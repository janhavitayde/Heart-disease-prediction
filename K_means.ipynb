```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples , silhouette_score
```

```python
data=pd.read_csv("/content/heart_disease_data.csv")
```

```python
data.head()
```

```python
#Basic information of data
data.info()
```

```python
plt.scatter(data.iloc[:,3],data.iloc[:,7])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Heart disease")
plt.show
```

```python
kMeans = KMeans(n_clusters=3)
```

```python
kMeans.fit(data.iloc[:,1:-1])
```

```python
print(kMeans.cluster_centers_)
```

```python
plt.scatter(data.iloc[:,3],data.iloc[:,7] ,c = kMeans.labels_,cmap='rainbow')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Heart Disease")
plt.show
```

```python
silhouette_avg = silhouette_score(data.iloc[:,1:-1],kMeans.labels_)
print(silhouette_avg)
```
