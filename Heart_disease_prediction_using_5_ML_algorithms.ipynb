```python
#let's transform categorical values into dummies/Convert categorical variable into dummy/indicator variables.

data = pd.get_dummies(data, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
#We will test some classification algorithms: Logistic regression, svm, stochastic gradient descent , decision tree, random forest.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
```

The dataset is split into two sets: a training set and a testing set. 70% of the dataset will be used as a training set and 30% will be used for testing.

```python
y = data['target']
X = data.drop('target',axis=1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
LR_classifier = LogisticRegression(random_state=0)
clf = svm.SVC()
forest=RandomForestClassifier(n_estimators=20, random_state=12,max_depth=6)
treee = tree.DecisionTreeClassifier(criterion = 'entropy',random_state=0,max_depth = 6)
GBC=GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0)
NB=GaussianNB()
```

```python
LR_classifier.fit(X_train, y_train)
clf.fit(X_train, y_train)
treee.fit(X_train, y_train)
forest.fit(X_train, y_train)
GBC.fit(X_train, y_train)
NB.fit(X_train, y_train)
```

```python
print(forest)
```

```python
print(GBC)
```

```python
#traing accuracy
y_pred=LR_classifier.predict(X_train)
y_predsvm=clf.predict(X_train)
y_predtree=treee.predict(X_train)
y_predforest=forest.predict(X_train)
y_predGBC=GBC.predict(X_train)
y_predNB=NB.predict(X_train)
```

```python
print(accuracy_score(y_train, y_pred))
print(accuracy_score(y_train, y_predsvm))
print(accuracy_score(y_train, y_predtree))
print(accuracy_score(y_train, y_predforest))
print(accuracy_score(y_train, y_predGBC))
print(accuracy_score(y_train, y_predNB))
```

```python
#test accuracy
y_pred=LR_classifier.predict(X_test)
y_predsvm=clf.predict(X_test)
y_predtree=treee.predict(X_test)
y_predforest=forest.predict(X_test)
y_predGBC=GBC.predict(X_test)
y_predNB=NB.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
print(accuracy_score(y_test, y_predsvm))
print(accuracy_score(y_test, y_predtree))
print(accuracy_score(y_test, y_predforest))
print(accuracy_score(y_test, y_predGBC))
print(accuracy_score(y_test, y_predNB))
```

Printing all accuracies

```python
print("ACCURACY")
print("")
print("Logistic Regression:", accuracy_score(y_test, y_pred)*100,"%")
print("Support Vector Machine:",accuracy_score(y_test, y_predsvm)*100,"%")
print("Decision Tree:",accuracy_score(y_test, y_predtree)*100,"%")
print("Random Forest:",accuracy_score(y_test, y_predforest)*100,"%")
print("Gradient Boosting:",accuracy_score(y_test, y_predGBC)*100,"%")
print("Gaussian Naive Bayes:",accuracy_score(y_test, y_predNB)*100,"%")
```
